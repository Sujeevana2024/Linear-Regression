# Assumption-1 Residuals are Normally Distributed

```python
y_pred = model.predict(xc)
y1 = y.copy()
y1['y_pred'] = y_pred
y1['resid'] = y1['price'] - y1['y_pred']
y1.head()
```

```python
resid = model.resid
resid
```

```python
plt.figure(figsize = (10,4))
sns.distplot(resid)
plt.show()
```

```python
sns.displot(resid,kde=True)
```

## Q-Q plot

```python
from scipy import stats
stats.probplot(resid,plot=plt)
plt.show()
```

## Test for Normality
## jarque bera test

```python
print(stats.jarque_bera(resid))
```

## Assumption-3: There is no heteroscedasticity    OR Residuals are Homoscedasticity

```python
y = df[['price']]
x = df.drop(['price','INDUS','AGE'],axis=1)
xc = sm.add_constant(x)
```

```python
model = sm.OLS(y,xc).fit()
y_pred = model.predict(xc)
```

```python
resid = model.resid
```

```python
fig,ax = plt.subplots(figsize=(10,5))
sns.regplot(x=y_pred,y=resid,lowess=True,ax=ax)
plt.show()
## lowess --> locally weighted sum of squares.
```

#### Goldfeld Test.

```python
import statsmodels.stats.api as sms
name = ['F statistic','p-value']
test = sms.het_goldfeldquandt(model.resid,xc)
# test = sms.het_goldfeldquandt(model.resid,model.model.exog)
test
```

### Assumption-4: Residuals are not auto-correlated

```python
y = df[['price']]
x = df.drop('price',axis=1)
xc = sm.add_constant(x)

model = sm.OLS(y,xc).fit()
y_pred = model.predict(xc)
resid = model.resid
model.summary()
```
